```python
marks=np.array([10,20,30])
print(marks)
print(marks.ndim)
```

```python
#Flaten
#copy of original and give a one d array
s=np.array([1,2,3,4,5,6])
new_arr=s.flatten()
print(new_arr)
print(s)
```

```python
#Ravel-Donot create copy give a one-d array
s=np.array([[1,2,3,4,5,6],[1,2,3,4,5,6]])
new_arr=s.ravel()
print(new_arr)
```
