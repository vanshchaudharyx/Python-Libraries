**Pandas**<br>
Two Data structures<br>
Data frame & Data Series

```python
import pandas as pd
```

```python
print(pd.__version__)
```

```python
# Series- Only one coloumn
k=pd.Series([10,20,30,40,50])
print(k)
#String,INT,Float,Boolean These are the data types on which we work in that library
```

```python
k=pd.Series([10,20,30,40,50])
index=['a101','a102','a103','a104','a105']
k.index=index #Customize indexing in series.
print(k)
print(k['a101'])
print(k['a101':'a104'])
k['a101']=100
print(k)
```

```python
k=pd.Series([10,20,30,40,50])
k[k<40]=50 #Conditionals in pd.
print(k)
```

**Data Frame**

```python
student={
    "name":['A','B','C'],
     "marks":[20,21,22],
    "age":[21,24,21]

}
df=pd.DataFrame(student)
print(df)
```

```python
result={
    "name":['A','B','C'],
     "marks":[20,21,22]

}
df=pd.DataFrame(result)
print(df)
```

```python
#by list
student=[
    ["A","B","C"],
    [20,21,22,'Rudra']
]
df=pd.DataFrame(student)
print(df)
```

```python
import numpy as np
data=np.array([[10,20],[30,40],[50,60]])
df=pd.DataFrame(data)
print(df)
```

```python
#Empty dataframe
df=pd.DataFrame()
print(df.empty)
print(df)
```

```python
#Attributes of dataframes.
student={
    "name":['A','B','C'],
     "marks":[20,21,22],
    "age":[21,24,21]

}
df=pd.DataFrame(student)
print(df.shape)
print(df.size)
print(df.ndim)
print(df.columns)
print(df.index)
print(df.dtypes)
print(df.values)
print(df.empty)
```

```python
student={
    "name":['A','B','C'],
     "marks":[20,21,22],
    "age":[21,24,21]

}
df=pd.DataFrame(student)
print(df.marks)
print(df.name)
print(df.age)
print(df[["marks","age"]]) #Column Selection
marks=df[['marks']]
print(type(marks))
```

```python
student={
    "name":['A','B','C'],
     "marks":[20,21,22],
    "age":[21,24,21]

}
df=pd.DataFrame(student)
#Variation of Slicing
print(df[0:2])
print(df[0:4])
print(df[-1:])
print(df[::2])
print(df[::-1])
print(df[:3])

```

**Named Labeling**

```python
student={
    "name":['A','B','C'],
     "marks":[20,21,22],
    "age":[21,24,21]

}
df=pd.DataFrame(student)
df.index=['a','b','c']
print(df)
print(df.loc[["a","c"]])
print(df['a':'c'])
print(df['c':])
#In label based indexing last indexed element is included..
print(df['-b':])


```

**iloc**

```python
student={
    "name":['A','B','C'],
     "marks":[20,21,22],
    "age":[21,24,21]

}
df=pd.DataFrame(student)
print(df.iloc[1])
print(df.iloc[1,2])
print(df.iloc[1:2])

```

**HEAD**

```python
student={
    "name":['A','B','C'],
     "marks":[20,21,22],
    "age":[21,24,21]

}
df=pd.DataFrame(student)
print(df.head) #5 is the default value of head.
print(df.tail(2)) #5 is the default value of head.
```

```python
student={
    "name":['A','B','C'],
     "marks":[20,21,22],
    "age":[21,24,32]

}
df=pd.DataFrame(student)
print(df.info())
```

```python
#Describe
student={
    "name":['A','B','C'],
     "marks":[20,21,22],
    "age":[21,24,32]

}
df=pd.DataFrame(student)
print(df.describe())

```
