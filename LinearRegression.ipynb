```python
import pandas as pd
from sklearn.linear_model import LinearRegression
```

sklearn-> sckitlearn


```python
data ={
    'Area':[1000,1200,1500,1800,2000,2500],
    'Price':[50,60,75,90,100,125]
}
```

```python
df=pd.DataFrame(data)
```

```python
X=df[["Area"]]
Y=df['Price']
```

```python
model=LinearRegression()
```

```python
model.fit(X,Y)
```

```python
new_house_area=[[2200]]
predicted_price=model.predict(new_house_area)

print("House Area :",new_house_area[0][0],"sq.ft")
print("Predicted House Price :",round(predicted_price[0],2),"Lakhs")
```

**House Price Prediction**

```python
house = {
    "Area":[1200,1500,1800,2000,2200,2500,2700,3000,3200,3500],
    "Bedrooms":[2,3,3,3,4,4,4,5,5,5],
    "Bathrooms":[2,2,3,3,3,4,4,4,5,5],
    "Age":[10,8,6,5,4,3,2,2,1,1],
    "Price":[45,58,72,80,90,105,115,130,145,160]
}

df = pd.DataFrame(house)
```

```python
X = df[["Area","Bedrooms","Bathrooms","Age"]]
y = df["Price"]
```

```python
model = LinearRegression()
```

```python
model.fit(X, y)
```

```python
new_house = pd.DataFrame({
    "Area":[2400],
    "Bedrooms":[4],
    "Bathrooms":[3],
    "Age":[3]
})

prediction = model.predict(new_house)

print("Predicted Price:", round(prediction[0],2), "Lakhs")
```
