**MATPLOTLIB**<br>


```python
import matplotlib.pyplot as plt
```

```python
from matplotlib.lines import lineStyles
#Line chart
x=[1,2,3,4,5]
marks=[60,40,30,46,90]
marks2=[29,67,64,12,89]
plt.plot(x,marks,color='pink',linewidth=5,marker='*',linestyle='--',markerfacecolor='red',label="std")
plt.plot(x,marks2,color='blue',linewidth=5,linestyle='--',marker='*',markerfacecolor='red',label="std")
plt.title("Line Chart")
plt.legend()
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

**Scatter Chart**

```python
hours=[1,2,3,4,5,6]
marks=[36,54,34,76,80,90]
marks2=[38,45,65,80,8,70]
color=['red','green','blue','yellow','pink','black']
plt.scatter(hours,marks,color=color,marker='o',s=234,edgecolors='black')
plt.scatter(hours,marks2,marker='o',s=234,edgecolors='black')
plt.title("Scatter Chart")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid(True)
plt.show()
```

**BAR CHART**

```python
std_name=["A","B","C","D","E"]
marks=[10,20,30,40,50]
color=['red','green','blue','yellow','pink','black']
plt.bar(std_name,marks,color=color,width=0.3,edgecolor='black',linewidth=3,alpha=0.5)
# plt.barh(std_name,marks,color=color,height=0.3,edgecolor='black',linewidth=3,alpha=0.5)
plt.title("Bar Chart")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

**Histogram Chart**



```python
import numpy as np
import matplotlib.pyplot as plt

# Generate 1000 random values
data = np.random.normal(
    loc=50,      # Mean
    scale=10,    # Standard deviation
    size=1000    # Number of data points
)

plt.hist(data, bins=20)


plt.xlabel("Values")
plt.ylabel("Frequency")
plt.title("Histogram")

plt.show()
```

**PIE CHART**

```python
course=['Java','AI','Python','C++','C']
std=[40,50,60,70,20]
colors=['red','green','blue','yellow','pink']
explode=[0,0,0.2,0.1,0]
plt.pie(std,labels=course,colors=colors,autopct='%1.2f%%',explode=explode,radius=1.5,wedgeprops={"edgecolor":"black","linewidth":2},textprops={"fontsize":15,"fontweight":"bold","color":"black"})
plt.title('Pie Chart')
plt.show()
```

```python
students=['Alice','B','C','D','E']
marks=[40,50,60,70,20]
colors=['red','green','blue','yellow','pink']
explode=[0,0,0.2,0.1,0.5]
plt.pie(marks,labels=students,colors=colors,autopct='%1.0f%%',explode=explode,radius=1.5,wedgeprops={"edgecolor":"black","linewidth":5},textprops={"fontsize":15,"fontweight":"bold","color":"black"})
plt.title('Pie Chart')
plt.show()
```

**AREA CHART**

```python
# area chart
days=[1,2,3,4,5,6,7]
sales=[100,150,120,180,200,170,220]
profit=[50,50,60,100,120,90,140]
expense=[50,70,60,80,80,80,80]
plt.fill_between(days,sales,color='green',alpha=0.3,label="sales")
plt.fill_between(days,expense,color='purple',alpha=0.5,label="expense")
plt.plot(days,expense,color='yellow',marker='o',linestyle='--')
plt.plot(days,sales,color='black',marker='o',linestyle='--')
plt.title("Area Chart")
plt.legend()
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

**SUB-PLOT CHART**

```python
x=[1,2,3,4,5]
y1=[10,20,15,30,25] #LINE
y2=[5,8,6,9,7]
y3=[15,20,25,6,10]#SCATTER
y4=[20,15,78,12,98]#AREA
plt.subplot(2,2,1)#row,coloumn chart
plt.plot(x,y1,color='red',marker='o')
plt.title("Line Chart")
plt.subplot(2,2,2)
plt.bar(x,y2,color='green')
plt.title("Bar Chart")
plt.subplot(2,2,3)
plt.scatter(x,y3,color='blue')
plt.title("Scatter Chart")
plt.subplot(2,2,4)
plt.plot(x,y4)
plt.title("Area Chart")
plt.tight_layout()
plt.show()

```

**Figure Cutomization**

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 20, 30, 40, 50]

plt.figure(
    figsize=(5, 5),
    dpi=100,
    facecolor="skyblue"
    ,edgecolor="red"
    ,linewidth=5
)

plt.plot(x, y)
plt.title("Figure Customization")
plt.show()
```

**AXIS CUSTOMIZATION**

```python
x = [1, 2, 3, 4, 5]
y = [10, 30, 20, 50, 40]
plt.figure(
    figsize=(5, 5),
    edgecolor="red",
    linewidth=5
)
plt.style.use("dark_background")

plt.plot(x,y)
plt.xlim(0,6)
plt.ylim(0,60)
plt.text(3,30,"Medium",color="red",fontweight="bold")
plt.text(3,60,"High",color="Black",fontweight="bold")
plt.text(3,19,"Low",color="Blue",fontweight="bold")
# plt.xticks([1,2,3,4,5])
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.title("Axis Customization")
plt.show()
```

**3D CHARTS**

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig=plt.figure(figsize=(10,10))
ax=fig.add_subplot(111,projection="3d")
x=[1,2,3,4,5]
y=[5,6,7,8,9]
z=[10,11,12,13,14]
ax.plot(x,y,z)
ax.set_title("3D Chart")
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.set_zlabel("Z-axis")
plt.show()
```

```python
fig=plt.figure(figsize=(10,10))
ax=fig.add_subplot(111,projection="3d")
x=[1,2,3,4,5]
y=[5,6,7,8,9]
z=[10,11,12,13,14]
ax.scatter(x,y,z,color="red")
ax.plot(x,y,z,color="red")
ax.set_title("3D Chart")
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.set_zlabel("Z-axis")
plt.show()
```
